```python
import pandas as pd
import numpy as np

from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import Perceptron
```

```python
df = pd.read_csv("cleaned.csv").iloc[:,1:].drop(columns=["kills","deaths"])
```

```python
df = pd.get_dummies(df)
```

```python
df.head()
```

```python
y = df["result"]
X = df.drop(columns = ["result"])
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
param_grid = {'solver': ["newton-cg", "lbfgs", "liblinear", "sag", "saga"]}
clf = GridSearchCV(LogisticRegression(), param_grid, n_jobs= -1)
clf.fit(X_train,y_train)
print("Best estimator found by grid search:")
print(clf.best_estimator_)
```

```python
print("The training error is:",1-clf.score(X_train, y_train))
print("The test error is:",1 -clf.score(X_test, y_test))
```

```python
ypred = clf.predict_proba(X)
ypred[::,1]
```

```python
df2 = pd.read_csv("cleaned.csv").iloc[:,1:]
```

```python
df2["win_ratio"] = ypred[::,1]* 100
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
df2
```

```python
param_grid = {'alpha': [0.00005,0.0001, 0.0005, 0.001, 0.005, 0.01, 0.05, 0.1, 0.5]}
clf1 = GridSearchCV(Perceptron(random_state=0), param_grid, n_jobs= -1)
clf1.fit(X_train,y_train)
print("Best estimator found by grid search:")
print(clf1.best_estimator_)
```

```python
print("The training error is:",1-clf1.score(X_train, y_train))
print("The test error is:",1 -clf1.score(X_test, y_test))
```

```python
mean_squared_error(clf1.predict(X),y)
```
